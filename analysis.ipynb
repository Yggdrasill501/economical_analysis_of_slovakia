### Gross Domestic Product per capita

```python
import pandas
import numpy
import matplotlib.pyplot as plt
```

```python
data = 'data/gdp/gdp.csv'
gdp_df= pandas.read_csv(data)
gdp_df
```

```python
plt.figure(figsize=(10,6))
plt.plot(gdp_df["Year"], gdp_df["GDP per Capita(USD)"], marker='o')
plt.title("GDP per Capita in Slovakia (2018-2022)")
plt.xlabel("Year")
plt.ylabel("GDP per Capita (in thousands)")
plt.grid(True)
plt.show()
```

```python
gdp_df= pandas.read_csv(data)
gdp_df
```

```python
gdp_df= pandas.read_csv(data)
gdp_df.set_index('Year', inplace=True)

fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(gdp_df.index, gdp_df['Anual Growth(%)'], color=['red' if x < 0 else 'green' for x in gdp_df['Anual Growth(%)']], alpha=0.6, label='Anual Growth(%))')

ax1.set_ylabel('Anual Growth(%)', color='black')
ax1.tick_params('y', colors='black')

ax2 = ax1.twinx()
ax2.plot(gdp_df.index, gdp_df['GDP per Capita(USD)'], color='blue', marker='o', label='GDP per Capita(USD)')
ax2.set_ylabel('GDP per Capita(USD)', color='blue')
ax2.tick_params('y', colors='blue')

plt.title('GDP per Capita and Annual Growth Rate in Slovakia (2018-2022)')
ax1.set_xlabel('Year')

fig.tight_layout()
plt.show()
```

```python
gdp_df['GDP Difference'] = gdp_df['GDP per Capita(USD)'].diff()
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.bar(gdp_df.index, gdp_df['Anual Growth(%)'], color=['red' if x < 0 else 'green' for x in gdp_df['Anual Growth(%)']], alpha=0.6, label='Anual Growth(%)')
ax1.set_ylabel('Anual Growth (%)', color='black')
ax1.tick_params('y', colors='black')

ax2 = ax1.twinx()
ax2.plot(gdp_df.index, gdp_df['GDP per Capita(USD)'], color='blue', marker='o', label='GDP per Capita(USD)')
ax2.set_ylabel('GDP per Capita(USD)', color='blue')
ax2.tick_params('y', colors='blue')

ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(gdp_df.index, gdp_df['GDP Difference'], color='purple', marker='x', linestyle='--', label='GDP Difference(USD)')
ax3.set_ylabel('GDP Difference(USD)', color='purple')
ax3.tick_params('y', colors='purple')

plt.title('GDP per Capita, Annual Growth, and Yearly Difference in Slovakia (2018-2022)')
ax1.set_xlabel('Year')

fig.tight_layout()
plt.show()
```
